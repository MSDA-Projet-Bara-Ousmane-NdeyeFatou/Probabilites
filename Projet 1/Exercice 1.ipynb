```python
import numpy as np
import matplotlib.pyplot as plt
from random import *
```

```python
"""------------------------------------------------------- 
1_ Tirage des réalisations suivant la loi uniforme [0, 10]
-------------------------------------------------------"""
n = 1
def tirage(n): 
    for i in range(1, n):
        valeurs = np.random.uniform(0,10,n)     
    return (valeurs)  
print("Tirage aléatoire suivant la loi uniforme dans l'intervalle [0, 10]: ", tirage(3))
```

```python
"""------------------------------------------------
2_ Calcul de la réalisation de la moyenne empirique
------------------------------------------------"""
def moyenne(n):
    moy = tirage(n).mean()
    return moy
print("La moyenne empirique de cette série est :", moyenne(3))
```

```python
"""------------------------------------
3_ Répétons 5000 fois les étapes 1 et 2 
------------------------------------"""

def realisations(N, n):   
    liste_5000_moyennes = []
    for i in range(N): 
        liste_5000_moyennes = liste_5000_moyennes + [moyenne(n)]
    return liste_5000_moyennes

print("Le nombre de réalisation de la moyenne empirique est : ", len(realisations(5000, 10)))
print("La liste des ", len(realisations(5000, 10)), " moyennes empiriques réalisées : ", realisations(5000, 10))
```

```python
"""----------------------------------------------- 
4_ Histogramme des 5000 réalisations de la moyenne
-----------------------------------------------"""

rls = realisations(5000, 10)
plt.hist(rls)
plt.title("Histogramme de  réalisations de la moyenne empirique")
plt.show()
```
