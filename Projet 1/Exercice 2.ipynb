```python
import numpy as np
import matplotlib.pyplot as plt
from random import *
```

```python
"""------------------------------------------------------- 
1_ Tirage des réalisations suivant la loi Khi-Deux
-------------------------------------------------------"""

def tirage(n): 
    for i in range(1, n):
        valeurs = np.random.chisquare(2, n)     
    return (valeurs)  
print("Tirage aléatoire suivant la loi du Khi-Deux :", tirage(3))

```

```python
"""------------------------------------------------
2_ Calcul de la réalisation de la moyenne empirique
------------------------------------------------"""

def moyenne(n):
    moy = tirage(n).mean()
    return moy
print("La moyenne empirique de cette série est :", moyenne(3))
```

```python
"""------------------------------
3_ Calcul de la réalisation de Zn
------------------------------"""

def ZedEn(n):
    moy = moyenne(n)
    fraction = (moy - 2) / 2
    racine = np.sqrt(n)
    zeden = fraction * racine
    return zeden
print(ZedEn(2))
```

```python
"""---------------------------------------
4_ Répétons 5000 fois les étapes 1, 2 et 3
---------------------------------------"""

def realisations(N, n):   
    liste_5000_zeden = []
    for i in range(N): 
        liste_5000_zeden = liste_5000_zeden + [ZedEn(n)]
    return liste_5000_zeden

print("Le nombre de réalisation de la variable Zn est : ", len(realisations(5000, 10)))
#print("La liste des ", len(realisations(5000, 10)), " variables Zn réalisées : ", realisations(5000, 10))
```

```python
"""--------------------------------------------------- 
5_ Histogramme des 5000 réalisations de la variable Zn
---------------------------------------------------"""

rls = realisations(5000, 10)
plt.hist(rls, density = True)
plt.title("Histogramme de 5000 réalisations de la variable Zn pour n = 10")
plt.show()
```
