```python
import string 
import urllib.request
import matplotlib.pyplot as plt
```

```python
"""-------------
Lecture du livre
-------------"""

x = urllib.request.urlopen("https://r-stat-sc-donnees.github.io/LesMiserables1.txt") 
livre=str(x.read()) 
print(livre)
```

```python
"""--------------------------------------------------------
Suppressions majuscules, ponctuations et division en jetons
--------------------------------------------------------"""

livre = livre.lower()
print(livre)
```

```python
livre = livre.split()
print(livre)
```

```python
"""---------------------------------------------------------
Calcul du nombre d'apparition des mots (un, et, le, il, est) 
---------------------------------------------------------"""

print("Le nombre d'apparition du mot 'un' est: ", livre.count("un"))
print("Le nombre d'apparition du mot 'et' est: ", livre.count("et"))
print("Le nombre d'apparition du mot 'le' est: ", livre.count("le"))
print("Le nombre d'apparition du mot 'il' est: ", livre.count("il"))
print("Le nombre d'apparition du mot 'est' : ", livre.count("est"))
```
